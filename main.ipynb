```python
from bs4 import BeautifulSoup
import requests

url = 'https://takeuforward.org/strivers-a2z-dsa-course/strivers-a2z-dsa-course-sheet-2/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the details of steps
all_steps = soup.find_all('summary', id=lambda x: x and x.startswith('dsa_sheets_problems_table_row_summary'))

# Iterate over each step
for step in all_steps:
    step_name = step.text.strip()
    print(f"Step: {step_name}")
    step=step+1
    step_name = step.text.strip()
    print(f"Step: {step_name}")



    # Find all the details of substeps within each step
    substeps_details_container = step.find_next('details')
    if substeps_details_container:
        substeps_table = substeps_details_container.find('table')
        if substeps_table:
            substeps_details = substeps_table.find_all('tr')

            # Iterate over each substep details
            for substep_detail in substeps_details:
                # Extract the substep name and associated links
                substep_name_tag = substep_detail.find('td')
                if substep_name_tag:
                    substep_name = substep_name_tag.text.strip()
                    links = substep_detail.find_all('a', href=True)
                    links_info = [(link.text.strip(), link['href']) for link in links]

                    # Print the substep name and associated links
                    # print(f"Substep: {substep_name}")
                    for link_info in links_info:
                        print(f"Problem: {link_info[0]} - Link: {link_info[1]}")
        else:
            print("No substeps found for this step.")
    else:
        print("No substep container found for this step.")
```
